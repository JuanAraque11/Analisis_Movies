Importar las libreriÃÅas necesarias, cargar el dataset y explorar el dataset.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

peliculas = pd.read_csv('https://raw.githubusercontent.com/JuanAraque11/Analisis_Movies/main/imdb_top_1000.csv')

# Mostrar los dos primeros registros
peliculas.head(2)
```

```python
print("Forma del dataset:")
print(peliculas.shape)
```

Datos faltantes en el dataset.

```python
peliculas.isna().sum()
```

Reemplazar los datos nulos del dataset para los datos faltantes.

```python
# Reemplazar valores faltantes de la columna 'Certificate' por la moda
peliculas['Certificate'] = peliculas['Certificate'].fillna(peliculas
 ['Certificate'].mode()[0])

# Reemplazar valores faltantes de la columna 'Meta_score' por la mediana
peliculas['Meta_score'] = peliculas['Meta_score'].fillna(peliculas
 ['Meta_score'].median())

# Reemplazar valores faltantes de la columna 'Gross' por la moda
peliculas['Gross'] = peliculas['Gross'].fillna(peliculas['Gross'].mode()[0])

# Mostrar el dataset actualizado
peliculas.isnull().sum()
```

Seperar los g√©neros de cada registro para crear graficos ded barras y ver la distribuci√≥n de los mismos.

```python
# Separar los g√©neros de cada registro
peliculas['Genre'] = peliculas['Genre'].str.split(', ')

# Crear una lista de todos los g√©neros
all_genres = [genre for sublist in peliculas['Genre'] for genre in sublist]

# Contar la frecuencia de cada g√©nero
genre_counts = pd.Series(all_genres).value_counts()

print(genre_counts)
```

Crear gr√°fico de barras.

```python
# Graficar el gr√°fico de barras
plt.figure(figsize=(10, 6))
genre_counts.plot(kind='bar', color='skyblue')
plt.title('Distribuci√≥n de Producciones por G√©nero')
plt.xlabel('G√©nero')
plt.ylabel('Cantidad de Producciones')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Ver el top 10 de peliculas m√°s votadas.

```python
# Crear un nuevo df ordenado descendientemente por la columna 'No_of_Votes'
top_5_peliculas = peliculas.sort_values(by='No_of_Votes',
                                        ascending=False).head(5)

# Graficar los primeros 10 registros
fig,ax=plt.subplots(figsize=(12,5))
sns.barplot(x=top_5_peliculas['Series_Title'], y=top_5_peliculas['No_of_Votes'],
            color='skyblue')
plt.title('Top 5 peliculas m√°s votadas', fontweight = 'bold', fontsize = 15)
plt.xlabel('Peliculas', fontsize = 10, fontweight = 'bold')
plt.ylabel('Votos', fontsize = 10, fontweight = 'bold')
plt.show()
```

Ver los directores con m√°s peliculas en el top.

```python
# Contar la cantidad de pel√≠culas por director
director_counts = peliculas['Director'].value_counts()

# Seleccionar los directores con m√°s pel√≠culas (por ejemplo, los 10 primeros)
top_directors = director_counts.head(10)

# Graficar los directores con m√°s pel√≠culas
plt.figure(figsize=(10, 6))
sns.barplot(x=top_directors.index, y=top_directors.values, color='lightblue')
plt.title('Directores con m√°s pel√≠culas en el conjunto de datos',
          fontweight='bold')
plt.xlabel('Director', fontsize=12, fontweight='bold')
plt.ylabel('Cantidad de pel√≠culas', fontsize=12, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

visualizar la proporci√≥n de producciones seg√∫n su certificaci√≥n.

```python
# Contar la frecuencia de cada certificaci√≥n
certificate_counts = peliculas['Certificate'].value_counts()

# Graficar el gr√°fico de pastel
plt.figure(figsize=(8, 8))
plt.pie(certificate_counts, labels=certificate_counts.index, autopct='%1.1f%%',
        startangle=140, colors=['lightcoral', 'lightskyblue', 'lightgreen'])
plt.title('Proporci√≥n de Producciones por Certificaci√≥n')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

Gr√°fico de Dispersi√≥n para Puntuaci√≥n de IMDb vs. Meta Score

```python
# Crear el gr√°fico de dispersi√≥n
plt.figure(figsize=(8, 6))
plt.scatter(peliculas['IMDB_Rating'], peliculas['Meta_score'], color='green',
            alpha=0.5)
plt.title('Puntuaci√≥n de IMDb vs. Meta Score')
plt.xlabel('IMDB Rating')
plt.ylabel('Meta Score')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Visualizar el n√∫mero de pel√≠culas por anÃÉo.

```python
# Contar la cantidad de pel√≠culas por a√±o
movies_per_year = peliculas['Released_Year'].value_counts().sort_index()

# Graficar la cantidad de pel√≠culas por a√±o
plt.figure(figsize=(18, 6))
sns.barplot(x=movies_per_year.index, y=movies_per_year.values,
            color='lightgreen')
plt.title('Cantidad de Pel√≠culas Realizadas por A√±o', fontweight='bold')
plt.xlabel('A√±o', fontsize=12, fontweight='bold')
plt.ylabel('Cantidad de Pel√≠culas', fontsize=12, fontweight='bold')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Identificar las 7 estrellas que m√°s veces aparecen en el top.

```python
# Crear una serie con todas las estrellas de las columnas 'Star1, 2, 3, 4'
all_stars = pd.concat([peliculas['Star1'], peliculas['Star2'],
                       peliculas['Star3'], peliculas['Star4']])

# Contar la cantidad de veces que aparece cada estrella
star_counts = all_stars.value_counts()

# Crear grafico de barras de las 7 primeras estrellas
plt.figure(figsize=(10, 6))
star_counts.head(7).plot(kind='bar', color='lightblue')
plt.title('Primeras 7 estrellas', fontweight='bold')
plt.xlabel('Estrella', fontsize=12, fontweight='bold')
plt.ylabel('Aparaciones', fontsize=12, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Ver el top 5 peliculas mejor puntuadas del genero animacion.

```python
# Filtrar las pel√≠culas con el g√©nero 'animation'
animation_movies = peliculas[peliculas['Genre'].apply(lambda x: 'Animation'
                                                      in x)]

# Ordenar las pel√≠culas por su calificaci√≥n IMDb de forma descendente
top_5_animation_movies = animation_movies.sort_values(by='IMDB_Rating',
                                                      ascending=False).head(5)

# Graficar las 5 pel√≠culas mejor rankeadas
plt.figure(figsize=(10, 6))
sns.barplot(x='IMDB_Rating', y='Series_Title', data=top_5_animation_movies)
plt.title('Top 5 Pel√≠culas de Animaci√≥n Mejor Rankeadas', fontweight='bold')
plt.xlabel('Calificaci√≥n IMDb', fontsize=12, fontweight='bold')
plt.ylabel('Pel√≠cula', fontsize=12, fontweight='bold')
plt.tight_layout()
plt.show()
```
